## Dropna and Get Dummies

<h3 style="text-align: left;" markdown="3">Dropna</h3>


```python
df=df.dropna()
nobs=len(df)
df.dtypesdf=df.dropna()
nobs=len(df)
df.dtypes
```

<h3 style="text-align: left;" markdown="3">Get dummies</h3>

```python
iForeign=pd.get_dummies(df['foreign'])['Foreign']
```

```python
iForeign
```

```python
df.iloc[:,-1]=iForeign
df
```

Now there is no problems in this dataset

## Regression

```python
y=df.iloc[:,2]
X=df.iloc[:,[1,3,4,5,6,7,8,9,10,11]]
```

```python
model = LinearRegression().fit(X,y)
```

Return $R^2$

```python
model.score(X,y)
```

Return intercept

```python
model.intercept_
```

Return coeffcient

```python
model.coef_
```

Return estimate $\hat{y}$ for each X

```python
model.predict(X)
```

```python
#another way#
model.intercept_ + np.sum(model.coef_ * X, axis=1)
```

## Get results just like stata
